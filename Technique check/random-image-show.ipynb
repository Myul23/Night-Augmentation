```python
import os
from glob import glob
import numpy as np
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data_path = "result"
if not os.path.exists(data_path):
    print("No data folder")
    exit()
```

```python
style_0 = np.random.choice(glob(f"{data_path}/*0/*"), 20, replace=False)
base = glob(f"{data_path}/*1/*")

style_1 = []
for address in style_0:
    address = address.split("\\")[-1]
    def findString(string):
        return address in string
    idx = list(map(findString, base)).index(True)
    style_1.append(base[idx])
```

```python
_, axes = plt.subplots(5, 8, figsize=(30, 22))
plt.subplots_adjust(hspace=0.01, wspace=0.04)

ax = axes.ravel()
for idx, (add_0, add_1) in enumerate(zip(style_0, style_1)):
    image_0 = cv2.imread(add_0)
    image_1 = cv2.imread(add_1)

    ax[idx * 2].imshow(cv2.cvtColor(image_0, cv2.COLOR_BGR2RGB))
    ax[idx * 2].axis("off")
    ax[idx * 2 + 1].imshow(cv2.cvtColor(image_1, cv2.COLOR_BGR2RGB))
    ax[idx * 2 + 1].axis("off")
```
